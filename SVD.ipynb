```python
def mc(R, J, lambda_, maxit, thresh):
    clf = SoftImpute(J=J, thresh=thresh, lambda_=lambda_, maxit=maxit)
    clf.fit(R)
    return(clf.predict(R))
```

```python
def cal_mse(R, J, lambda_, maxit, x_test, y_test, thresh=1e-09):
    R_columns = R.columns
    R_index = R.index
    if not isinstance(R, (np.ndarray, np.generic)):
        R = np.array(R)
    def mc(R, J, lambda_, maxit):
        clf = SoftImpute(J=J, thresh=thresh, lambda_=lambda_, maxit=maxit)
        clf.fit(R)
        return(clf.predict(R))
    R_hat = pd.DataFrame(mc(R, J, lambda_, maxit), columns = R_columns, index = R_index)
    predict_value_list = []
    for i, x in enumerate(x_test.values.tolist()):
        predict_value_list.append(R_hat.loc[x[0], x[1]])
    return(mean_squared_error(np.array(y_test), np.array(predict_value_list)))
    
```
